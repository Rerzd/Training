## Part 2: Cleaning Datasets:
Also called pre-procesing or Data wrangling, it is very important to know how to handle missing values, correct data formatting as well as standarize and normalize data in order to ensure the quality of the information

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
```

```python
plus_file="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
df = pd.read_csv(plus_file, header = None)
df.head()
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df = pd.read_csv(plus_file, names = headers)
df.head()
```

```python
#Convert "?" to Nan

df.replace ("?", np.nan, inplace = True)
df.head()
```

```python
#use .isnull() or .notnull() to identify missing values
missing_data = df.isnull()
missing_data.head()
```

```python
#Count missing values in each column based on the previous boolean values

for column in missing_data.columns.values.tolist():
  print(column)
  print(missing_data[column].value_counts())
  print("")
```

### Deal with missing data

in case you have missing data con can either

1. Drop Data by dropping the whole row or whole column

2. Replace data: It can be with the mean, frecuency or based on other functions

In case there are a lots of empty entries in the column, then you can drop it. In this case, It wouldn't be the best choice.

Let's appply each method for different columns

```python
#Replace by Mean. first we have to calculate the mean

avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print(f" The Average of normalized-losses is: {avg_norm_loss}")
```

```python
# Replace by Mean. Note: When coding, python jumped a warning to change  the following code since it will obsolete in the future
# df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)

df.replace({"normalized-losses":np.nan}, avg_norm_loss, inplace = True)
```

```python
#Calculate for bore
avg_bore = df["bore"].astype('float').mean(axis = 0)
print(f" The average of bore is: {avg_bore}")
```

```python
#Replace Nan

df.replace({"Bore":np.nan}, avg_bore, inplace = True)
```

```python
# Calculate for stroke

avg_stroke = df['stroke'].astype('float').mean(axis = 0)
print(f"  The average of stroke is: {avg_stroke}")
```

```python
#Replace Nan

df.replace({"stroke":np.nan}, avg_stroke, inplace = True)
```

```python
# Horsepower
avg_horsepower = df['horsepower'].astype('float').mean(axis = 0)
print(f" The Average horsepower is: {avg_horsepower}")

#Replace Nan

df.replace({"horsepower":np.nan}, avg_horsepower, inplace = True)
```

```python
# Peak-rpm

avg_peakrpm = df['peak-rpm'].astype('float').mean(axis=0)
print(f" The average peak_rpm", avg_peakrpm)

#Nan

df.replace({"peak-rpm":np.nan}, avg_peakrpm, inplace = True)
```

```python
#See which values are present ina  particular column with value.counts()

df['num-of-doors'].value_counts()
```

```python
# 4 is the most commmon type. we can use the .idxmax() metrod to calculate the most common type
df['num-of-doors'].value_counts().idxmax()
```

```python
#Replace the missing 'num-of-dorrs with the most frequent

df.replace({'num-of-dooes':np.nan}, 'four', inplace = True)
```

```python
#Drop the whole row of a car that does not have price
df.dropna(subset=["price"], axis = 0, inplace = True)

# Reset index since we dropped two rows
df.reset_index(drop=True, inplace = True)
```

```python
df.head(10)
```

### Data Normalization:

It is the process of trasnforming values of several variables into a similar range. Typically it includes

1. Scaling the variable so the variable average is 0
2. Scaling the variable so the variance is 1
3. Scaling the variable so the variable values range from 0 to 1

Example:

To demonstrate normalization, soy you want to scale the columns "length", "width" and "height"

**Target**: Normaliza those variables so theur value ranges from 0 to 1

**Approach**: Replace the original value by (og_val)/(max_val)


```python
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height'] = df['height']/df['height'].max()
```

## Part 2: Cleaning Datasets (Laptops example)



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
new_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-Coursera/laptop_pricing_dataset_mod1.csv"
df = pd.read_csv(new_path)
df.rename(columns={'Unnamed: 0':'id'}, inplace=True)
df.head()
```

```python
df.info()
```

```python
df[['Screen_Size_cm']] = np.round(df[['Screen_Size_cm']],2)
df.head()
```

```python
#Evaluate dataset for missing data
missing_data = df.isnull()
print(missing_data.head())
for column in missing_data.columns.values.tolist():
  print(column)
  print(missing_data[column].value_counts())
  print("")
```

```python
# We have to replace weight_kg with the mean for thsi columns since there are only 5 null values

Average_Weight_kg = df['Weight_kg'].astype('float').mean(axis=0)
print(f" The Average of the weight in kilograms is: {Average_Weight_kg}")

df.replace({"Weight_kg":np.nan}, Average_Weight_kg, inplace = True)
```

```python
# Replace Scree_Size_cm with the most frequent value

freq_Screen_Size_cm = df['Screen_Size_cm'].value_counts().idxmax()
print(f" The most frequent value of screen size is: {freq_Screen_Size_cm}")

df.replace({"Screen_Size_cm":np.nan}, freq_Screen_Size_cm, inplace = True)
```

```python
#Check that "Wight_kg" and "Screen_Size_cm" are indeed both float values
df.dtypes
```

```python
#data Standarization in both weight and screen.

#Screen size has an standard unit for inches, so we have to convert the column to inches
df['Screen_Size_cm'] = df['Screen_Size_cm']/2.54
df.rename(columns={'Screen_Size_cm': 'Screen_Size_in'}, inplace = True)

#Weight is in kg so we to change it to pounds
df['Weight_kg'] = df['Weight_kg'] * 2.205
df.rename(columns={'Weight_kg': 'Weight_pounds'}, inplace = True)
```

```python
#Data normalization: We have to apply this process to the CPU_frequency

df['CPU_frequency'] = df['CPU_frequency']/df['CPU_frequency'].max()
df['CPU_frequency'].head()
```

```python
#Binning: We have to categorize the pirce data in low, medium and high and store it in the new column "Price-binned"
```

```python
# Make a copy first
df['Price'] = df['Price'].astype(int, copy = True)

#Define the bins
bins = np.linspace(min(df['Price']), max(df['Price']), 4)

#Define the groups
group_names = ['Low', 'Medium', 'High']

#Make the bins

df['Price-binned'] = pd.cut(df['Price'], bins, labels = group_names, include_lowest = True)
df[['Price', 'Price-binned']].sample(15)
```

```python
#Now, to end this excercise lets make a indicator variables

#First, we check the columns and look for 'Screen'
df.columns
```

```python
#Now, we create a dummy of Screen
dummy_variable_1 = pd.get_dummies(df['Screen'])
dummy_variable_1.head()
```

```python
#Now we separate screen in this two variables

dummy_variable_1.rename(columns={'Full HD':'Screen_Full_HD', 'IPS Panel': 'Screen_IPS_Panel'}, inplace = True)
dummy_variable_1.head()
```

```python
#Now we concat the modified columns with the original dataframe and erase the original 'Screen'
df = pd.concat([df, dummy_variable_1], axis = 1)
df.drop('Screen', axis = 1, inplace = True)
```
